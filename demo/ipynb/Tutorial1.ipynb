# A Introduction to ILAMB Variables

This notebook is meant to explain and showcase the capabilities of the Variable object which is part of the ILAMB python package. 

First, I will import the functionality that I will need in this notebook. The first line of this import is only for this notebook version and causes the plots to render inline in the document. You would not need this in your own notebooks. If you are new to python, these lines you can skip over. Few people begin their scripts importing all the functions they will need. Rather, these lines get populated as you develop your scripts.

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from ILAMB.Variable import Variable,FromNetCDF4
from ILAMB.Post import ColorBar
from ILAMB.constants import mid_months,lbl_months
from netCDF4 import Dataset
import os
```

I will showcase the capabilities of the Variable object using the Gross Primary Production (gpp) data from the Fluxnet-MTE Global product. This data already exists in NetCDF4 files as part of the data download which comes along with the ILAMB package. So I simply create the variable from the file.

```python
gpp = FromNetCDF4(os.environ["ILAMB_ROOT"]+"DATA/gpp/FLUXNET-MTE/derived/gpp.nc","gpp")
```

This routine pulls as much information as it can out of the NetCDF4 file. For example, we can take a look at the name and unit:

```python
print "name = %s, unit = %s" % (gpp.name,gpp.unit)
```

If the variable is defined over space and/or time, this information is also loaded into the Variable object automatically. In this case, the gpp data are montly means of the flux-rate which is defined over the globe from 1982 to 2008. So we can check that the data was in fact understood to be spatially and temporally defined.

```python
print "is temporal = %s" % gpp.temporal
print "is spatial = %s" % gpp.spatial
```

The various routines of the Variable object behave differently depending on what is defined internally inside. For example, if the Variable is temporal, then we can integrate it over the time period:

```python
time_int_gpp = gpp.integrateInTime()
```

This returns another instance of a Variable object, but it has now been integrated in time:

```python
print "name = ", time_int_gpp.name
print "unit = ", time_int_gpp.unit
print "is temporal = ", time_int_gpp.temporal
print "is spatial = ", time_int_gpp.spatial
```

Note that a few changes have occurred automatically. First the name of the variable is created by postpending the gpp variable's name with "_integrated_over_time". This is important as the variable objects themselves track how they were created. Also note that the unit has now lost its time dimension and that the result is no longer temporal.

Sometimes instead of an integral, we want rather to have the mean over the time period. This can be accomplished by including the "mean" keyword when integrating:

```python
time_int_gpp = gpp.integrateInTime(mean=True)
print "name = ", time_int_gpp.name
print "unit = ", time_int_gpp.unit
```

Note that the variable name also reflects that this is now a mean over the time period and the unit has its temporal component restored, but now appears in days. This is because the temporal dimension is internally stored in days and thus the units are converted internally to match. However, these numbers could be quite large and it is common to look at this mean data in terms of grams per square meter per day. There is some rudimentary support in the Variable object for basic unit conversions.

```python
time_int_gpp = time_int_gpp.convert("g m-2 d-1")
print "name = ", time_int_gpp.name
print "unit = ", time_int_gpp.unit
```

The Variable objects can also be plotted if they are spatial or temporal using matplotlib. The spatial plots make use of the Basemap toolkit. To make a plot, simply pass a matplotlib Axes object to the plot routine.

```python
fig,ax = plt.subplots()
time_int_gpp.plot(ax)
plt.show()
```

This will render a basic plot of the variable in what I deem is an asthetic default. However, the plot routine supports many keywords which changes the look of the plot. For example, we can change the [colormap](http://matplotlib.org/examples/color/colormaps_reference.html) to "Greens" and the plotting region to "nhaf" (Northern Hemisphere Africa).

```python
fig,ax = plt.subplots()
time_int_gpp.plot(ax,
                  region = "nhaf",
                  cmap   = "Greens")
plt.show()
```

While I plan to expand the supported keywords to include common option requests, the user does not need to go through this plotting routine. The data inside the variable object may be accessed directly (no data hiding) and then Basemap can be used directly inside of user scripts. The main idea of the Variable.plot() routine is to give a user a quick way to render output in an asthetic way.

One notable absence from my plot is the colorbar. For technical reasons relating to how matplotlib plots colorbars, I have opted to keep their creation separate from the plotting routine. 

```python
# make two axes on a 10x1 grid, the globe takes 9 rows
# and the colorbar takes 1 row
ax0 = plt.subplot2grid((10, 1), (0, 0),rowspan=9) 
ax1 = plt.subplot2grid((10, 1), (9, 0)) 
time_int_gpp.plot(ax0,
                  region = "nhaf",
                  cmap   = "Greens")
ColorBar(ax1,
         vmin  = time_int_gpp.data.min(),
         vmax  = time_int_gpp.data.max(),
         cmap  = "Greens",
         label = time_int_gpp.unit)
plt.show()
```

Notice that in order to define the min and max on the colorbar, I directly accessed the data inside the ILAMB Variable object. This is important to keep in mind because while the Variable object is designed to make analysis operations easy and transparent, you are certainly free to access data directly and do what you wish.

Taking the temporal integral gives us an idea of the spatial distribution of gpp over the globe. Alternatively, we could take spatial integrals of the original gpp data which will reveal more information about the cycle and overall trend over the time period. Here we will perform the integration, convert the unit, and plot the result.

```python
space_int_gpp = gpp.integrateInSpace().convert("Pg y-1")
print "name = ", space_int_gpp.name
print "unit = ", space_int_gpp.unit
fig,ax = plt.subplots()
space_int_gpp.plot(ax,color="g")
ax.set_xlabel("Year")
ax.set_ylabel("Mean GPP [%s]" % space_int_gpp.unit)
plt.show()
```

Notice that we used the same plot routine as before, but this time the variable is defined only over time and not space. The plot routine then automatically understands the kind of plot you need and places it on the axis that you pass into it. 

The spatial integration can optionally be done over regions by using the "region" keyword. So here we show the same spatial integral but this time just over the amazon. The name reflects that the integration was over a specific region. Unsurprisingly we see that the cycle amplitude is far smaller as the tropical regions stay more active year round.

```python
amazon_int_gpp = gpp.integrateInSpace(region="amazon").convert("Pg y-1")
print "name = ", amazon_int_gpp.name
print "unit = ", amazon_int_gpp.unit
fig,ax = plt.subplots()
amazon_int_gpp.plot(ax,color="g")
ax.set_xlabel("Year")
ax.set_ylabel("Mean GPP [%s]" % amazon_int_gpp.unit)
plt.show()
```

It is clear from these plots that there is an annual cycle to this data. The ILAMB Variable object also supports extracting this information. Notice that as before, the name of the variable is automatically generated to reflect the history of what has happened to get to this point in the analysis. In addition to a new Variable object, the annual cycle routine returns the mean day of peak gpp and its standard deviation. In this case the output reflects that the mean peak day of gpp is the 259th day of the year (0 indexing) which corresponds to September. While we are skeptical that there is really no variation in that mean, we realize that the input data was monthly mean gpp. This means that whatever variation surely exists does so at a time scale beyond our resolution.

```python
cycle_gpp,tmax_mean,tmax_std = space_int_gpp.annualCycle()
print "name = ", cycle_gpp.name
print "unit = ", cycle_gpp.unit
print "tmax = %f +- %f [d]" % (tmax_mean,tmax_std)
```

We can also plot the annual cycle using the plot routine. In this case, the annual cycle routine also stored the standard deviation of the mean gpp at each month and the plot routine knows to render this data if it is available. The shaded regions of the below plot reflect plus and minus one standard deviation relative to the mean, shown as a solid line. 

```python
fig,ax = plt.subplots()
cycle_gpp.plot(ax,color="g")
ax.set_xlim(1850,1851)
ax.set_xticks(1850+mid_months/365.)
ax.set_xticklabels(lbl_months)
ax.set_ylabel(cycle_gpp.unit)
plt.show()
```

In order to facilitate communication of results, the ILAMB Variable object also supports a direct dump into a NetCDF4 Dataset. You can simply open a dataset and then pass it into the toNetCDF4() routine. All of the spatial and temporal grids will get automatically entered, but with checks to avoid duplication. This is also a great way to perform analysis using this package and then other NetCDF4-friendly tools to do custom visualizations that you may have already built.

```python
filename = os.environ["HOME"] + "/results.nc"
# have to remove the file, won't let me overwrite
os.system("rm -f %s" % filename)
dataset = Dataset(filename,mode="w")
time_int_gpp .toNetCDF4(dataset)
space_int_gpp.toNetCDF4(dataset)
cycle_gpp    .toNetCDF4(dataset)

# what I have to do to print a dump in this notebook
print os.popen("ncdump -h %s" % filename).read()
```
